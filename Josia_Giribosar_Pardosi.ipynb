#Forecasting 4 month Customer Lifetime Value

```python
orders_summ.head()
```

```python
#embuat ringkasan statistik untuk penjualan pelanggan dalam interval 4 bulan menggunakan data dari orders_summ DataFrame. 
#clv_freq diatur sebagai '4M' yang berarti bahwa semua faktur pelanggan akan dikelompokkan ke dalam siklus 4 bulan. 
#Data frame data_df dibuat dengan mengelompokkan data orders_summ berdasarkan CustomerID dan InvoiceDate menggunakan metode groupby. 
#Kemudian, agregasi dihitung untuk kolom Sales menggunakan fungsi sum, groupby_mean, dan groupby_count.
#Kolom data diberi nama dengan bergabung menggunakan _ pada data_df.columns.
clv_freq = '4M'
data_df = orders_summ.reset_index().groupby(['CustomerID', pd.Grouper(key='InvoiceDate', freq=clv_freq)]).agg({
    'Sales': [sum, groupby_mean, groupby_count],})

data_df.columns = ['_'.join(col).lower() for col in data_df.columns]
data_df = data_df.reset_index()
data_df
```

```python
#menambahkan informasi "Recency" atau kebaruan pembelian terakhir pelanggan ke dalam DataFrame data_df. 
#Proses ini dimulai dengan membuat sebuah dictionary date_month_map yang akan digunakan untuk mengonversi tanggal pembelian menjadi bulan. 
#Dictionary ini dibuat dengan memanggil fungsi enumerate() pada objek tanggal unik di DataFrame data_df, yang kemudian diurutkan secara terbalik. 
#Selanjutnya, key  dan nilai dari dictionary dibentuk dengan menggunakan format string M_%s untuk setiap bulan yang ditemukan dalam objek tanggal.
date_month_map = {str(x)[:10]: 'M_%s' % (i+1) for i, x in enumerate(sorted(data_df.reset_index()['InvoiceDate'].unique(), reverse=True))}
data_df['M'] = data_df['InvoiceDate'].apply(lambda x: date_month_map[str(x)[:10]])
data_df.head()
```

```python
data_df['M'].unique()
```

```python
#summary statistik dari penjualan pelanggan pada setiap 4 bulan terakhir. 
#Pertama-tama, kita menggunakan metode pivot_table dari pandas untuk membuat tabel pivot yang akan menampilkan 3 nilai dari setiap 4 bulan terakhir, 
#yaitu sales_sum, sales_avg, dan sales_count.
features_df = pd.pivot_table(data_df.loc[data_df['M'] != 'M_1'], values=['sales_sum', 'sales_avg', 'sales_count'], columns='M', index='CustomerID')

features_df.columns = ['_'.join(col) for col in features_df.columns]

features_df.head()
```

```python
#DataFrame ini dibuat dengan mengambil data dari DataFrame data_df yang telah dibuat sebelumnya 
#Hanya memilih data dengan nilai kolom 'M' sama dengan 'M_1'.
response_df = data_df.loc[data_df['M'] == 'M_1', ['CustomerID', 'sales_sum']]

response_df.columns = ['CustomerID', 'CLV_'+clv_freq]
response_df
```

```python
#Menampilkan hasil dataframe yang terdiri dari fitur-fitur transaksi pelanggan selama 4 bulan dan nilai customer lifetime value pada bulan ke-5.
sample_set_df = features_df.merge(
    response_df, 
    left_index=True, 
    right_on='CustomerID',
    how='left'
)

sample_set_df = sample_set_df.fillna(0)
sample_set_df
```
